**Que 1) Write a python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.**

```python
import  requests
from  bs4 import BeautifulSoup

url = requests.get('https://en.wikipedia.org/wiki/Main_Page')
soup = BeautifulSoup(url.text, 'html.parser')
story = soup.find_all(['h1', 'h2','h3'])

# Printing header tags
for i in story:
    print(i)
```

**Que:2) Write a python program to display IMDB’s Top rated 100 movies’ data(i.e. Name, IMDB rating, duration, director, star cast - 10 main characters)**

```python
# USING SELENIUM
from selenium import webdriver
url="https://www.imdb.com/chart/top/?ref_=nv_mv_250"

driver=webdriver.Chrome(executable_path='chromedriver')
driver.get(url) # this command will search the given url in the browser
title=[]
rating=[]
director=[]
duration=[]
cast_name=[]
data={}
for i in range(1,101):
    driver.implicitly_wait(5)
    element=driver.find_element_by_xpath(f"//tbody[@class='lister-list']//tr[{i}]//td[@class='titleColumn']//a")
    element.click()
    ttl=driver.find_element_by_xpath("//div[@class='title_wrapper']//h1").text
    title.append(ttl)
    
    rat=driver.find_element_by_xpath("//div[@class='ratingValue']//span").text
    rating.append(rat)
    
    dirct=driver.find_element_by_xpath("//div[@class='credit_summary_item']//a").text
    director.append(dirct)
    
    dur=driver.find_element_by_xpath("//time").text
    duration.append(dur)
    
    try:
        casts=[]
        for j in range(2,11):
            cast=driver.find_element_by_xpath(f"//table[@class='cast_list']//tr[{j}]").text
            casts.append(cast)
    except:
        pass
    cast_name.append(casts)  
    driver.back()
    data={'Movie Name':title,'IMDB Rating':rating,'Director':director,'Duration':duration,'Star Cast':cast_name}

# Printing movie details in a dataframe
import pandas as pd
df_imdb=pd.DataFrame(data)
df_imdb
```

**Que 3)Write a python program to find geospatial coordinates (i.e. latitudes and longitudes) of a city name entered by user from ‘www.google.co.in/maps’.**

```python
from selenium import webdriver
from selenium.webdriver import ActionChains
url="https://www.google.co.in/maps/"

driver=webdriver.Chrome(executable_path='chromedriver')
driver.get(url)

# Accessing search bar
element=driver.find_element_by_xpath("//div[@id='gs_lc50']//input")
element.click()

# Sending data
element.send_keys("New Delhi")
element.send_keys("\n")

k=driver.find_element_by_xpath("//div[@class='widget-scene']")
ActionChains.context_click(k).perform()

# Accessing menu to get the coordinates
driver.find_element_by_xpath("//ul[@role='menu']//li[3]").click()
k=driver.find_element_by_xpath("//div[@class='widget-reveal-card-container']//button[2]").text
print("Coordinates are:",k)
```

**Que 4) Write a python program to scrap book name, author name, genre and book review of 5 books from ‘www.bookpage.com’**

```python
from selenium import webdriver
url="https://bookpage.com/reviews"

driver=webdriver.Chrome(executable_path='chromedriver')
driver.get(url)
book_details={}
author=[]
reviews=[]
genre=[]
book_name=[]

# for loop which will get all the details of the book
for i in range(1,6):
    review=[]
    name=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//h4//a").text
    book_name.append(name)
    
    auth=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//p").text
    author.append(auth)
    
    gen=driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//p[2]//a").text
    genre.append(gen)
    
    driver.find_element_by_xpath(f"//div[@class='row-fluid article-row'][{i}]//div[@class='read-full']//a").click() 
    driver.implicitly_wait(5)
    for j in range(2,5):
        rev=driver.find_element_by_xpath(f"//div[@class='article-body']//p[{j}]").text
        review.append(rev)
    driver.back()
    reviews.append(review)
book_details={"Book Name":book_name,"Author Name":author,"Genre":genre,"Review":reviews}

# Storing book data in a dataframe
import pandas as pd
df_book=pd.DataFrame(book_details)
df_book
```
